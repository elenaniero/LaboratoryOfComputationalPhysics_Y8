1\. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
import random

random_numbers = [random.randint(1, 100) for _ in range(20)]

file_name = "simple_data.txt"
with open(file_name, "w") as f:
    for number in random_numbers:
        f.write(f"{number}\n")

print("Data saved to simple_data.txt")
```

2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
matrix = np.random.rand(5, 5)

with open("data.txt", "w") as f:
    for row in matrix:
        f.write(" ".join(f"{val:.4f}" for val in row) + "\n")
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
data = []
with open("data.txt", "r") as f:
    for line in f:
        # Strip: remove the intial and final whitespaces of a row
        #split: make a list with words separated by a comma
        row = line.strip().split()
        data.append(row)

# Write the CSV file by hand
with open("data.csv", "w") as f:
    for row in data:
        # Join using commas instead of spaces
        f.write(",".join(row) + "\n")

print("Data saved to data.csv")
    
```

4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
def decode_credit_cards(path):
    with open("/home/elena/LaboratoryOfComputationalPhysics_Y8/credit_card.dat" , "rb") as f:
        data = f.read()

    bitstring = ""
    # Convert file bytes into a continuous bit string
    for byte in data:
        bitstring += f"{byte:08b}"

    i = 0
    current_line = ""
    cards = []

    # Process bits in chunks of 10 (6 bits + 4 padding)
    while i + 10 <= len(bitstring):
        char_bits = bitstring[i:i+6]        # first 6 bits = character
        padding   = bitstring[i+6:i+10]     # last 4 bits = padding (ignored)

        i += 10

        value = int(char_bits, 2)           # convert 6-bit value to number
        c = chr(value)                      # convert to character

        if c == "\n":                       # newline marks end of card number
            cards.append(current_line)
            current_line = ""
        else:
            current_line += c

    # Append last line if not empty
    if current_line:
        cards.append(current_line)

    return cards


# Example usage:
cards = decode_credit_cards("/home/elena/LaboratoryOfComputationalPhysics_Y8/credit_card.dat")
for c in cards:
    print(c)
```
