### Numpy basics

1\. Find the row, column and overall means for the following matrix:

```python
m = np.arange(12).reshape((3,4))
```

````python
import numpy as np

m = np.arange(12).reshape((3,4))

for i in range(3):
    r = m[i, : ]
    print(r)
    print("The mean of raw number ", i, "is: ", r.mean())
for j in range(4):
    c = m[ : ,j]
    print(c)
    print("The mean of column number ", j, "is: ", c.mean())

print("The overall mean is: ", m.mean())

````

````python
import numpy as np

m = np.arange(12).reshape((3,4))

print("The means of raws are: ", m.mean(axis=1))
print("The means of columns are: ", m.mean(axis=0))
print("The overall mean is: ", m.mean())
````

2\. Find the outer product of the following two vecotrs

```python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
```

Do this in the following ways:

   * Using the function outer in numpy
   * Using a nested for loop or list comprehension
   * Using numpy broadcasting operatoins


````python
import numpy as np

u = np.array([1,3,5,7])
v = np.array([2,4,6,8])

print("The outer product of the 2 arrays with the outer function in numpy is: \n", np.outer(u,v))

p = np.ones((len(u), len(v)))
for i in range(len(u)):
    for j in range(len(v)):
        p[i,j] = u[i]*v[j]

print("The outer product of the 2 arrays with the nested for loop is: \n", p)

b = np.tile(u, (len(v),1)).T
print("The outer product of the 2 arrays with the broadcasting operations is: \n", b*v)
````

4\. Use np.linspace to create an array of 100 numbers between 0 and 2π (includsive).

  * Extract every 10th element using slice notation
  * Reverse the array using slice notation
  * Extract elements where the absolute difference between the sine and cosine functions evaluated at that element is less than 0.1
  * Make a plot showing the sin and cos functions and indicate where they are close

````python
import numpy as np
import math 
import matplotlib.pyplot as plt

m = np.linspace(0, 2*math.pi,100)
extract = m[::10]
reverse = m[::-1]
mask = (abs(np.sin(m)-np.cos(m)) < 0.1)
filtered_m = m[mask]

print(m)
print(extract)
print(reverse)
print(filtered_m)
sine = np.array([(np.sin(i)) for i in filtered_m])
cos = np.array([(np.cos(i)) for i in filtered_m])

print(sine)
print(cos)
for i in range(len(sine)):
    plt.plot(m, np.sin(m), 'ro', m, np.cos(m), 'go', [sine[i] ,cos[i]], '-')
````

5\. Create a matrix that shows the 10 by 10 multiplication table.

 * Find the trace of the matrix
 * Extract the anto-diagonal (this should be ```array([10, 18, 24, 28, 30, 30, 28, 24, 18, 10])```)
 * Extract the diagnoal offset by 1 upwards (this should be ```array([ 2,  6, 12, 20, 30, 42, 56, 72, 90])```)

````python
import numpy as np

# Create 10×10 multiplication table
table = np.arange(1, 11) * np.arange(1, 11).reshape(-1, 1)

print("Multiplication Table:\n", table)

# Trace of the matrix
trace_val = np.trace(table)
print("\nTrace:", trace_val)

# Anti-diagonal (top-right → bottom-left)
anti_diag = np.fliplr(table).diagonal()
print("\nAnti-diagonal:", anti_diag)

# Diagonal offset by +1 (one above main diagonal)
diag_offset = np.diag(table, k=1) #np.diag(v, k): v is the array, if it is a vector (1D) the command construct a matrix with that vector in diagonal,
# if it is a matrix (2D) the command extract the diagonal of the matrix. 
# k is 0 by default, to extract the diagonal above the main one we have to set k>0, to extract the diagonal under the main one we have to set k<0.

print("\nDiagonal offset by 1 upwards:", diag_offset)
````

6\. Use broadcasting to create a grid of distances

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles
The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Construct a 2D grid of distances among each city along Route 66
  * Convert that in km (those savages...)

````python
import numpy as np

# Distances in miles from Chicago for each city
miles = np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])

miles_reshape = miles.reshape(10,1)
distances = np.abs(miles - miles_reshape)

print("Distance grid (miles):\n", distances)
distances_km = distances * 1.609

print("Distance grid (km):\n", distances_km)
````
